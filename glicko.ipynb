```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# expected outcome based on glicko system http://www.glicko.net/glicko/glicko.pdf
rd = 40
q = np.log(10)/400
g = 1/np.sqrt(1+3*q**2*(rd**2)/np.pi**2)
```

```python
r_diff = np.arange(-500,500)
```

```python
E = 1.0/(1+10**((g*r_diff)/400))
```

```python
d2 = (q**2*g**2*E*(1-E))**-1
```

```python
r_win = q/(1.0/rd**2+1.0/d2)*g*(1-E)
r_lose = q/(1.0/rd**2+1.0/d2)*g*(-E)
```

```python
plt.plot(r_diff,E)
```

```python
plt.plot(r_diff,d2)
```

```python
plt.plot(r_diff,r_win)
plt.grid(True)
```

```python
plt.plot(r_diff,r_lose)
plt.grid(True)
```

```python
plt.plot(r_win,r_lose)
```

```python
plt.plot(r_diff,q/(1.0/rd**2+1.0/d2))
```
